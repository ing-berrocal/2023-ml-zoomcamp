## Se prepara la data para ser homogenia

```python
df.columns = df.columns.str.lower().str.replace(' ','_')
```

```python
for c in df.dtypes[df.dtypes=='object'].index.to_list():
    df[c] = df[c].str.lower().str.replace(' ','_')
```

```python
df.head().T
```

```python
tc = pd.to_numeric(df.totalcharges, errors='coerce')
```

```python
df[tc.isnull()][['customerid','totalcharges']]
```

```python
df.totalcharges = pd.to_numeric(df.totalcharges, errors='coerce')
df.totalcharges = df.totalcharges.fillna(0)
```

```python
df['churn'] = (df.churn == 'yes').astype(int)
```

```python
from sklearn.feature_extraction import DictVectorizer
```

```python
train_dicts = df_train[number_columns+categorycal_columns].to_dict(orient='records')
```

```python
dv = DictVectorizer(sparse=False)
```

### One hot encoding

```python
dv.fit(train_dicts)
```

```python
dv.get_feature_names_out()
```

```python
X_train = dv.transform(train_dicts)
```

```python
X_train.shape
```

```python
dv.get_feature_names_out()
```

```python
from sklearn.linear_model import LogisticRegression
```

```python
model = LogisticRegression(solver='liblinear', random_state=1)
model.fit(X_train, y_train)
```

```python
val_dict = df_val[categorycal_columns + number_columns].to_dict(orient='records')
X_val = dv.transform(val_dict)
```

```python
y_val_pred = model.predict(X_val)
```

```python
model.predict_proba(X_val)
```

```python
(y_val == y_val_pred).mean()
```
